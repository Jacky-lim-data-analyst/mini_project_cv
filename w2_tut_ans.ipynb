## Question 2

````python
def is_even_or_odd(num):
    """Function to check whether the integer is odd or even."""
    if isinstance(num, int):
        if num % 2 == 0 and num != 0:
            print(f"{num} is an even number.")
        elif num % 2 == 1:
            print(f"{num} is an odd number.")
        else:
            print(num, " not even nor odd")
    else:
        print(num, " is not integer.")
````

````python
is_even_or_odd(2.15)
is_even_or_odd(46)
is_even_or_odd(-101)
is_even_or_odd(0)
````

### Question 3

#### Python list array operation (elementwise multiplication)

````python
%timeit [i*2 for i in list(range(1_000_000))]
````

#### Equivalent Operation with Numpy

````python
%timeit 2 * np.arange(1_000_000)
````

We can see that NumPy operation is 10x faster than native Python list.

### Question 5

````python
black = np.zeros((200, 200), dtype=np.uint8)

h, w = black.shape
xc, yc = w // 2, h // 2
# define radius and execute function over coordinates of numpy array
r = 25
distance = np.fromfunction(lambda i, j: np.sqrt((i - xc)**2 + (j - yc)**2), (200, 200))

black[distance <= 25] = 255
````

````python
plt.imshow(black, cmap=plt.cm.gray, vmin=0, vmax=255)
plt.xticks([]), plt.yticks([])
plt.show()
````
